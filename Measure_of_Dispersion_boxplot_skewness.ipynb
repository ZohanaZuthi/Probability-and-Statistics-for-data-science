# **Measure of Dispersion**

Dispersion means how the data is spread.
### **What is Variance in Statistics?**  

Variance (**œÉ¬≤** or **s¬≤**) is a measure of how **spread out** the data points are around the **mean**. It quantifies the **dispersion** or variability in a dataset. A **high variance** means data points are widely spread, while a **low variance** means they are closely packed around the mean.

---

Population Variance (œÉ¬≤):
œÉ¬≤ = ( Œ£ (xi - Œº)¬≤ ) / N

Sample Variance (s¬≤):
s¬≤ = ( Œ£ (xi - xÃÑ)¬≤ ) / (n - 1)

Where:
xi  = Each data point
Œº   = Mean of the population
xÃÑ   = Mean of the sample
N   = Total number of data points in the population
n   = Total number of data points in the sample
Œ£   = Summation notation


---

### **Example Calculation**
**Dataset:** `[2, 4, 6, 8, 10]`  
1. **Mean:**  
\[
\bar{x} = \frac{2+4+6+8+10}{5} = 6
\]
2. **Variance Calculation:**
\[
\sigma^2 = \frac{(2-6)^2 + (4-6)^2 + (6-6)^2 + (8-6)^2 + (10-6)^2}{5}
\]
\[
= \frac{16 + 4 + 0 + 4 + 16}{5} = \frac{40}{5} = 8
\]

üìå **Variance = 8** ‚úÖ

---

### **Python Code to Compute Variance**
Using **NumPy**:
```python
import numpy as np

data = [2, 4, 6, 8, 10]

# Population variance
pop_variance = np.var(data)

# Sample variance
sample_variance = np.var(data, ddof=1)

print("Population Variance:", pop_variance)
print("Sample Variance:", sample_variance)
```

## 1. Percentiles:

A percentile is a value below which a given percentage of observations fall. For example, the 50th percentile means that 50% of the data points are less than or equal to that value.

- Formula for Percentiles:  
  To find the percentile P_k (where k is the percentile you‚Äôre interested in), you would order the data from smallest to largest, and then use the formula:

  P_k = (k / 100) * (n + 1)

  where:
  - k is the desired percentile (e.g., 25, 50, 75).
  - n is the number of data points.

Real-World Example:
Consider exam scores for a class of 100 students. If you are interested in the 80th percentile score, it means you're looking for the score below which 80% of the students scored.

Real-World Implication:
- In job applications, a candidate in the 90th percentile for a test score is considered to have scored better than 90% of all applicants. Percentiles are used in fields like education, sports, marketing, and health for comparative analysis.

##2. Quartiles:

Quartiles are special percentiles that divide data into four equal parts. There are three quartiles:
- Q1 (First Quartile): 25th percentile (25% of the data is below this value).
- Q2 (Second Quartile): 50th percentile (this is the median of the data).
- Q3 (Third Quartile): 75th percentile (75% of the data is below this value).

Formula for Quartiles:
- Q1: It is the median of the lower half of the data (not including the overall median if the number of data points is odd).
- Q2: This is the median of the entire dataset.
- Q3: It is the median of the upper half of the data.

Outliers are usually defined as values that are below or above a certain threshold from the interquartile range (IQR). The IQR is the range between Q1 (the first quartile) and Q3 (the third quartile).

IQR = Q3 - Q1

Real-World Implication:
- Business: In sales, the first quartile might represent the lower 25% of your products in terms of performance, while the third quartile represents the top 25% of performers.
- Healthcare: In measuring patient recovery times, quartiles can help determine which 25% of patients took the longest or shortest recovery time.

##3. Box Plot:

A box plot (also known as a box-and-whisker plot) is a graphical representation of the quartiles and percentiles. It visually shows the minimum, first quartile, median, third quartile, and maximum values in the data.

- Box plot elements:
  - The Box: Represents the interquartile range (IQR), which is the distance between Q1 and Q3.
  - The Whiskers: Represent the range of the data, extending to the minimum and maximum values within a specified range (usually 1.5 * IQR).
  - The Line in the Box: This represents the median (Q2).

Box Plot with Curve:

A box plot can be overlaid with a density curve or histogram to show the distribution of data. This provides insight into whether the data is symmetric, skewed, or has outliers.

- Symmetric Distribution: The box and whiskers are balanced, and the curve in the histogram or density plot looks bell-shaped (e.g., normal distribution).
- Skewed Distribution: If the whiskers or the histogram curve leans to one side, it indicates skewness (positive or negative).
- Outliers: Any points outside the whiskers in the box plot are considered outliers.

Box Plot Example:

Here‚Äôs how a box plot could look:

  |---|----------------------|--------|---|
  Min  Q1        Median (Q2)    Q3      Max

- Min and Max: Represent the lowest and highest values in the data set.
- Q1, Median (Q2), Q3: Represent the quartiles as explained earlier.

Real-World Insights:

1. Education: Percentiles and quartiles are used in standardized testing to evaluate a student's performance relative to others. For example, if you are in the 90th percentile, it means you scored better than 90% of other students.
  
2. Sports: Percentiles can help compare an athlete's performance to their peers. For example, in a 100-meter race, being in the 95th percentile means you‚Äôre faster than 95% of the competitors.

3. Finance: Quartiles are used in portfolio management. Investors may use quartiles to understand the distribution of returns across different investments, helping them make more informed decisions.

4. Healthcare: Quartiles can be used in understanding recovery times, blood pressure ranges, or medical metrics to classify patients into risk categories (low, medium, high risk).

In Summary:
- Percentiles divide data into 100 equal parts.
- Quartiles divide data into four parts (25%, 50%, 75%, and 100%).
- These measures are useful in various fields to make sense of large datasets, understand distributions, and make comparative decisions.


# **Skewness:**

Skewness refers to the asymmetry or lack of symmetry in the distribution of data. It measures the extent to which the distribution deviates from a normal distribution. If a distribution is skewed, it means the data is not evenly spread out around the mean.

#### **Types of Skewness:**
1. **Positive Skew (Right Skew):**
   - The right tail (larger values) of the distribution is longer or fatter than the left tail.
   - The majority of data points are concentrated on the left side (lower values).
   - **mean** > **median** >**mode**
   
2. **Negative Skew (Left Skew):**
   - The left tail (smaller values) of the distribution is longer or fatter than the right tail.
   - The majority of data points are concentrated on the right side (higher values).
   - The **mean** is less than the **median** in negatively skewed data.

3. **No Skew (Symmetrical Distribution):**
   - The data is evenly distributed around the mean, forming a bell-shaped curve (normal distribution).
   - **mean** < **median** <**mode**

---

###Skewness Formula:

Skewness = (n / ((n - 1) * (n - 2))) * Œ£((x_i - xÃÑ) / œÉ)^3

Where:
- n = Number of data points.
- x_i = Each data point.
- xÃÑ = Mean of the data.
- œÉ = Standard deviation of the data.

Alternatively, a simplified formula for skewness is:

Skewness = (mean - median) / standard deviation

---





````python

import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.stats import skew

# Example data: Simulate positive skewed data (right skew)
np.random.seed(10)
data_positive_skew = np.random.exponential(scale=2, size=1000)

# Example data: Simulate negative skewed data (left skew)
data_negative_skew = -np.random.exponential(scale=2, size=1000)

# Calculate skewness
skew_positive = skew(data_positive_skew)
skew_negative = skew(data_negative_skew)

# Plot the distributions
plt.figure(figsize=(12, 6))

# Positive Skew (Right Skew)
plt.subplot(1, 2, 1)
sns.histplot(data_positive_skew, kde=True, color='blue')
plt.title(f"Positive Skew (Skewness = {skew_positive:.2f})")

# Negative Skew (Left Skew)
plt.subplot(1, 2, 2)
sns.histplot(data_negative_skew, kde=True, color='red')
plt.title(f"Negative Skew (Skewness = {skew_negative:.2f})")



plt.tight_layout()
plt.show()

````

### **Visual Explanation:**
- **Positive Skew (Right Skew):** The histogram of data with positive skew will show that the right side (larger values) has a long tail, while most of the data points are concentrated on the left side.
- **Negative Skew (Left Skew):** The histogram of data with negative skew will show that the left side (smaller values) has a long tail, while most of the data points are concentrated on the right side.

### **Real-World Insights:**

#### **1. Finance:**
   - **Stock Market Returns:** Stock market returns can often exhibit **positive skew**, where a few stocks may perform extraordinarily well, creating a longer right tail. Investors typically prefer assets with less skewness (normal distribution) because the performance is more predictable.

#### **2. Income Distribution:**
   - **Income:** In most countries, income data is positively skewed. A small number of people earn significantly higher incomes than the majority. This creates a right-skewed distribution where most people earn lower to average incomes, while a few earn very high incomes.
   
#### **3. Education:**
   - **Exam Scores:** In many cases, exam scores might show **negative skew** if most students perform well, but a few students perform poorly. This is because the majority of students cluster near the top, with fewer students scoring low.

#### **4. Real Estate:**
   - **Housing Prices:** Housing prices in most cities tend to exhibit **positive skew**. A few high-value properties (luxury homes) create a long right tail in the distribution of prices, while the majority of homes may fall within a moderate price range.

#### **5. Healthcare:**
   - **Disease Spread:** In some cases, the spread of diseases may show **positive skew**. Most people may have mild symptoms, but a few individuals may experience severe outcomes, creating a long tail in the distribution of severity.

### **Conclusion:**
Skewness provides valuable insights into the distribution of data and can help to make better predictions, analyze risk, and understand the behavior of datasets. Identifying skewness helps to choose the right statistical techniques, especially when making assumptions about the normality of data.

- Positive skew: Right tail, mean > median.
- Negative skew: Left tail, mean < median.
- No skew (normal): Symmetric distribution, mean = median.
