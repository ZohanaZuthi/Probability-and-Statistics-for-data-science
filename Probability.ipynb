### **P-Value in Statistics**  

The **p-value** (probability value) is a key concept in hypothesis testing that helps determine the significance of results.  

P value is the probability of a certain event.

## **3. Example**  
A researcher wants to test whether a new drug improves recovery time compared to the current treatment.  

- **Null Hypothesis (\(H_0\))**: The new drug has no effect.  
- **Alternative Hypothesis (\(H_1\))**: The new drug reduces recovery time.  
- After analysis, the **p-value = 0.03**.  
- Since \( p < 0.05 \), we **reject \(H_0\)** and conclude that the drug is effective.  

---

## **4. P-Value and Hypothesis Testing Steps**  
1. **State the hypotheses** (\(H_0\) and \(H_1\)).  
2. **Select a significance level (\(\alpha\))** (commonly 0.05).  
3. **Perform a statistical test** (e.g., t-test, z-test).  
4. **Calculate the p-value**.  
5. **Compare the p-value with \(\alpha\)**:  
   - If \( p < \alpha \), reject \(H_0\).  
   - If \( p \geq \alpha \), fail to reject \(H_0\).  

---

## **5. Common Statistical Tests that Use P-Values**  
- **t-Test**: Comparing means of two groups.  
- **ANOVA**: Comparing means of multiple groups.  
- **Chi-Square Test**: Checking independence between categorical variables.  
- **Regression Analysis**: Checking significance of predictors.
